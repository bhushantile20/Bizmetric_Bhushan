```python


# 1. Write a list comprehension to generate squares of numbers from 1 to 10.
square = [i**2 for i in range(1,11)]
print(square)
```

```python
# 2. Create a list of even numbers between 1 and 50 using list comprehension.
even = [i for i in range(1,51) if i%2==0]
print(even)
```

```python
# 3. Convert all strings in a list to uppercase using list comprehension.
words = ['hello', 'world', 'python']
upper_words = [w.upper() for w in words]
```

```python
print(upper_words)
```

```python
# 4. Given a list of integers, create a new list that contains only the positive numbers.
numbers = [-1, 2, -3, 4, -5, 6, 0]
positives = [n for n in numbers if n > 0]
print(positives)
```

```python
# 5. Create a list of tuples (num, num^2) for numbers 1 to 5.
num = [(i, i**2) for i in range(1,6)]
print(num)
```

```python
# 6. Extract all vowels from a given string using list comprehension.
text = "hello world"
vowels = [char for char in text if char in 'aeiouAEIOU']
print(vowels)
```

```python
# 7. Flatten a 2D list using list comprehension.
matrix = [[1,2,3], [4,5,6], [7,8,9]]
list = [item for row in matrix for item in row]
print(list)
```

```python
# 8. Replace all negative numbers in a list with 0 using list comprehension.
nums = [-1, 2, -3, 4, -5]
no_negative = [n if n >= 0 else 0 for n in nums]
print(no_negative)
```

```python
# 9. Given a list of words, create a list of lengths of each word.
word_list = ['cat', 'dog', 'elephant']
length = [len(w) for w in word_list]
print(length)
```

```python
# 10. Filter out words that start with the letter 'A' or 'a'.
words_list = ['apple', 'banana', 'ant', 'cat']
a_words = [w for w in words_list if w.lower().startswith('a')]
print(a_words)
```

```python
# 11. From a list of numbers, generate a list of "even" or "odd" strings using list comprehension.
num_list = [1,2,3,4,5]
even_odd = ['even' if n%2==0 else 'odd' for n in num_list]
print(even_odd)
```

```python

# 12. Create a list of numbers divisible by both 3 and 5 in range 1–100.
multiples = [i for i in range(1,101) if i%3==0 and i%5==0]
print(multiples)
```

```python
# 13. Write a nested list comprehension to generate a multiplication table for 1–5.
mult_table = [[i*j for j in range(1,6)] for i in range(1,6)]
print(mult_table)
```

```python
# 14. Convert a dictionary's keys into a list using list comprehension.
data_dict = {'a':1, 'b':2, 'c':3}
dict = [k for k in data_dict]
print(dict)
```

```python
# 15. Extract numeric digits from a string using list comprehension.
mixed = "abc12def345"
digits_only = [char for char in mixed if char.isdigit()]
print(digits_only)
```

```python
# 16. Use list comprehension to remove all spaces from a string.
spaced_text = "hello world python"
no_spaces = [char for char in spaced_text if char != ' ']
print(''.join(no_spaces))
```

```python
# 17. Create a list of characters that appear more than once in a string.
duplicate_text = "hello"
dup_chars = [char for char in set(duplicate_text) if duplicate_text.count(char) > 1]
print(dup_chars)
```

```python
# 18. From a list of sentences, generate a list of all words (split using list comprehension).
sentences_list = ["hello world", "python coding"]
all_words= [word for sentence in sentences_list for word in sentence.split()]
print(all_words)
```

```python
# 19. Create a list of unique elements from a list using list comprehension + condition.
repeated_nums = [1,2,2,3,1,4]
unique_nums = []
[unique_nums.append(x) for x in repeated_nums if x not in unique_nums]
print(unique_nums)
```

```python
# 20. Generate all pairs (x, y) where x is from list A and y is from list B (cartesian product).
list_a = [1, 2]
list_b = ['a', 'b']
all_pairs = [(x, y) for x in list_a for y in list_b]
print(all_pairs)
```
